Extracting top 3 cast

```python
def extract_cast3(cast_obj):
    counter = 0
    cast_list = []
    for i in ast.literal_eval(cast_obj):
        if(counter<3):
            cast_list.append(i['name'])
            counter+=1
        else:
            break
    
    return cast_list
    
```

```python
movie_dataset['cast'] = movie_dataset['cast'].apply(extract_cast3)
```

```python
def extarct_director_from_crew(crew_obj):
    director = []
    for i in ast.literal_eval(crew_obj):
        if(i['job'] == 'Director'):
            director.append(i['name'])
            break
    
    return director
```

Now, we will combine overview, keywords, genres and Director into one super columns with all details

```python
# Removing spaces between words of keywords, genres and cast so as to distinct between values
movie_dataset['genres'] = movie_dataset['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movie_dataset['genres'].value_counts()
```

```python
movie_dataset['keywords'] = movie_dataset['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movie_dataset['cast'] = movie_dataset['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movie_dataset['Director'] = movie_dataset['Director'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# Converting a str to a list for proper concatenation
movie_dataset['overview'] = movie_dataset['overview'].apply(lambda x:x.split())
```

```python
movie_dataset['tags'] = movie_dataset['overview'] + movie_dataset['genres'] + movie_dataset['keywords'] + movie_dataset['cast'] + movie_dataset['Director']
new_movie_dataset = movie_dataset[['movie_id','title','tags']]
```

```python
# Converting the list of tags again to a string
new_movie_dataset['tags'] = new_movie_dataset['tags'].apply(lambda x: " ".join(x))
```

```python
new_movie_dataset['tags'] = new_movie_dataset['tags'].apply(lambda x: x.lower())
```

Vectorization of tags data using "Bag of Words" stratergy

```python
new_movie_dataset['tags'][2]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vector = CountVectorizer(max_features=5000, stop_words='english')

cv = count_vector.fit_transform(new_movie_dataset['tags'])

print(type(cv))
```

```python
count_vector
```

```python
vectors = cv.toarray()
vectors
```

```python
print(type(count_vector.get_feature_names_out()))
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stemmer(text):
    base_word = []
    for i in text.split():
        base_word.append(ps.stem(i))
    
    return " ".join(base_word)
```

```python
print(ps.stem('loving'))
print(ps.stem('loves'))
print(ps.stem('loved'))

stemmer(new_movie_dataset['tags'][0])
```

```python
new_movie_dataset['tags'] = new_movie_dataset['tags'].apply(stemmer)
```

```python
feature_names = count_vector.get_feature_names_out()
print(len(feature_names))
print(" ".join(feature_names))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similiar_matrix = cosine_similarity(vectors)
```

```python
similiar_matrix
```

```python
sorted(similiar_matrix[0],reverse=True)[1:6]
```

Main Function

```python
def recommend(input_movie):
    movie_index = new_movie_dataset[new_movie_dataset['title'] == input_movie].index[0]
    similiarity_values = similiar_matrix[movie_index]
    movie_index_list = sorted(list(enumerate(similiarity_values)),reverse=True,key= lambda x:x[1])[1:6]

    for i in movie_index_list:
        print(new_movie_dataset.iloc[i[0]].title)
        
```
